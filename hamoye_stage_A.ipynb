## Question 11
What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

Hint: Perform a groupby sum aggregation on ‘Item’

```python
quiz_data.groupby('Item')[['Y2014', 'Y2017']].sum()
```

```python
animal_fat_2014 = quiz_data.groupby('Item')[['Y2014', 'Y2017']].sum().loc['Animal fats'][0]

animal_fat_2017 = quiz_data.groupby('Item')[['Y2014', 'Y2017']].sum().loc['Animal fats'][1]

print(f'Sum of animal fats in 2014: {round(animal_fat_2014, 2)}')

print(f'Sum of animal fats in 2017: {round(animal_fat_2017, 2)}')
```

## Question 13

What is the total number and percentage of missing data in 2016 to 2 decimal places?

```python
null_2016 = quiz_data['Y2016'].isnull().sum()
```

```python
total_2016 = len(quiz_data['Y2016'])
```

```python
null_percent = (null_2016/total_2016)*100
```

```python
print(f'Total missing data: {null_2016} \nPercent of missing data: {round(null_percent, 2)}')
```
